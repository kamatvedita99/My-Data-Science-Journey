# i a) Reading the dataset

```python
cereal_data = pd.read_csv("cereal.csv")
cereal_data.head()
```

Fields in the dataset:

Name: Name of cereal
mfr: Manufacturer of cereal
A = American Home Food Products;
G = General Mills
K = Kelloggs
N = Nabisco
P = Post
Q = Quaker Oats
R = Ralston Purina
type:
cold
hot
calories: calories per serving
protein: grams of protein
fat: grams of fat
sodium: milligrams of sodium
fiber: grams of dietary fiber
carbo: grams of complex carbohydrates
sugars: grams of sugars
potass: milligrams of potassium
vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
shelf: display shelf (1, 2, or 3, counting from the floor)
weight: weight in ounces of one serving
cups: number of cups in one serving
rating: a rating of the cereals (Possibly from Consumer Reports?)

# ii) mean, median, mode, range, variance of every column


```python
print("\n----------- Calculate Mean -----------\n")
print(cereal_data.mean())
```

```python
print("\n----------- Calculate Median -----------\n")
print(cereal_data.median())
```

```python
print("\n----------- Calculate Mode -----------\n")
mode_data=cereal_data.iloc[:,1:15].mode()
mode_data.fillna('-')
```

```python
print("\n----------- Calculate Range -----------\n")
cereal_data_non_cat = cereal_data.drop(columns=['name','mfr','type'])

print(cereal_data_non_cat.max()-cereal_data_non_cat.min())
```

```python
print("\n----------- Calculate Variance -----------\n")
cereal_data.var()
```
